DATA CLEANING

```python
# Drop rows with null values in FINAL PROGRAM ASSIGNMENT column
pd_data_cleaned = pd_data.dropna(subset=['FINAL PROGRAM ASSIGNMENT'])
len(pd_data_cleaned)
```

```python
def clean_imputate_data(column, default_values):
    data_counts = {}
    # Clean the selected column
    for index in pd_data_cleaned.index:
        # Get data without any trailing spaces, and other unnecessary character
        data = pd_data_cleaned.loc[index, column]
        if isinstance(data, str):  # Check if data is a string
            data = data.replace('\xa0', '').strip().replace(' ', ' ').rstrip()
        # Check if data is in default values
        if data in default_values or pd.isna(data):
            pd_data_cleaned.loc[index, column] = None
        else:
            data_counts[data] = data_counts.get(data, 0) + 1
        pd_data_cleaned.loc[index, column] = data
    # Get the most frequent data
    data_mode = max(data_counts, key=data_counts.get) if data_counts else None
    # Imputate null data with the most frequent data
    for index in pd_data_cleaned.index:
        data = pd_data_cleaned.loc[index, column]
        if not data or pd.isna(data):
            pd_data_cleaned.loc[index, column] = data_mode
```
